Finding the linear regression

```python
import matplotlib.pyplot as plt
from scipy import stats
#slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(data1[0],data1[1])
print("y = %6.2f x +%6.2f" % (slope, intercept))
```

```python
slope2, intercept2, r_value2, p_value2, std_err2 = stats.linregress(data2[0],data2[1])
print("y = %6.2f x +%6.2f" % (slope2, intercept2))
```

```python
slope3, intercept3, r_value3, p_value3, std_err3 = stats.linregress(data3[0],data3[1])
print("y = %6.2f x +%6.2f" % (slope3, intercept3))
```

```python
slope4, intercept4, r_value4, p_value4, std_err4 = stats.linregress(data4[0],data4[1])
print("y = %6.2f x +%6.2f" % (slope4, intercept4))
```

```python
data = pd.read_csv("Police_Department_Incident_Reports__Historical_2003_to_May_2018.csv")
```

```python
crime_cat_uni = data.Category.unique()
print(crime_cat_uni)
len(crime_cat_uni)
```

```python
import seaborn as sns
sns.countplot(x = 'Category',
              data = data,
              order = data['Category'].value_counts().index)
plt.xticks(rotation='vertical')
plt.show()
```
